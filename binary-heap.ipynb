## Basic Binary Heap

```python
from typing import List

class MaxBinaryHeap:
    
    size: int = 0
    pq: List[int] = []
    
    def swim(self, k: int):
        pos: int = k

        while pos > 1:
            j = round(pos / 2)
            if self.less(j, pos):
                self.exch(pos, j)
                pos = j
            else:
                break
    
    def sink(self, k: int):
        pos = k
        while pos < self.size:
            j = pos * 2
            if j > self.size:
                break
            if j < self.size and self.less(j, j+1):
                j += 1
            if not self.less(pos, j):
                break
            self.exch(pos, j)
            pos = j
    
    def less(self, parentK: int, k: int):
        return self.pq[parentK - 1] < self.pq[k - 1]
     
    def exch(self, parentK: int, k: int):
        tmp = self.pq[parentK - 1]
        self.pq[parentK - 1] = self.pq[k - 1]
        self.pq[k - 1] = tmp
    
    def insert(self, x: int):
        self.pq.append(x)
        self.size += 1
        self.swim(self.size)
    
    def delMax(self):
        max = self.pq[0]
        self.exch(1, self.size)
        self.pq.pop()
        self.size -= 1
        self.sink(1)
        return max
    
heap = MaxBinaryHeap()

heap.insert(1)
print(heap.pq)
heap.insert(3)
print(heap.pq)
heap.insert(19)
print(heap.pq)
heap.insert(29)
print(heap.pq)
heap.insert(2)
print(heap.pq)
heap.insert(6)
print(heap.pq)
heap.insert(4)
print(heap.pq)
heap.insert(61)
print(heap.pq)
heap.delMax()
print(heap.pq)
heap.delMax()
print(heap.pq)
heap.delMax()
print(heap.pq)
heap.delMax()
print(heap.pq)
heap.delMax()
print(heap.pq)
heap.delMax()
print(heap.pq)
heap.delMax()

print(heap.pq)
```
